**Question2:** During which month did the highest number of flights occur? Months are represented by numerical codes, with January corresponding to 1, February to 2, and so forth.

a) January <br>
b) May <br>

```python
# Count the occurrences of each month
month_counts = df['Month'].value_counts()

# Get the month with the highest count
highest_month = month_counts.idxmax()

print(f"The month with the highest number of flights is: {highest_month}")

# Optional: Map the numerical month code back to the month name for better readability
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

highest_month_name = month_names.get(highest_month, "Unknown Month")
print(f"This corresponds to the month: {highest_month_name}")
```

**Question 5:** Convert the values of 'Duration' into seconds. Enter the average duration (in seconds) of a flight. Enter your answer correct to two decimal places.

```python
# function to convert duration to seconds
def duration_to_seconds(time_num):
  t = time_num.split(' ')
  time_sec = 0
  for i in range(len(t)):
    if 'h' in t[i]:
      time_sec += int(t[i].replace('h',''))*3600
    elif 'm' in t[i]:
      time_sec += int(t[i].replace('m',''))*60
    elif 's'in t[i]:
      time_sec +=int(t[i].replace('s',''))
  return time_sec
```

```python
df['Duration'] = df['Duration'].apply(duration_to_seconds)
```

```python
round(df['Duration'].mean(),2)
```

**Question 6:** How many flights started in the Morning and arrived the destination at Evening?

```python
def time_cat(time_de):
  hour = int(time_de.split(':')[0])
  if hour >=5 and hour <12:
    return 'Morning'
  elif hour>=12 and hour<17:
    return 'Afternoon'
  elif hour>=17 and hour<20:
    return 'Evening'
  else:
    return 'Night'
```

```python
df['Dep_Time'] = df['Dep_Time'].apply(time_cat)
df['Arrival_Time'] = df['Arrival_Time'].apply(time_cat)
df_mor_even = df[(df['Dep_Time']=='Morning') & (df['Arrival_Time']=='Evening')]
print('Number of flights started in the Morning and arrived the destination at Evening is:',len(df_mor_even))
```
