````python
# import important libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
````

Common Instructions


---

```

- Load the wine dataset from sklearn.

- Split the dataset into train and test set with 70:30 ratio with random_state = 1

- Use this for all the questions that follow

```

````python
from sklearn.datasets import load_wine
wine = load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df['target'] = wine.target
# Display the few sample rows
df.sample(4)
````

````python
X = df.drop(columns=['target'])
y = df['target']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
````

Common instruction for Question 1 and 2


---


```
Train a Logistic Regression model with the following parameters:

max_iter = 10000

C as 0.5

random_state = 1

solver as 'sag'

Compute the accuracy on the test data.
```
## Question 1: What is the accuracy of the model on the test data? [Enter 3 decimal places]

````python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 10000, C=0.5, random_state = 1, solver='sag')
lr.fit(X_train,y_train)
````

````python
from sklearn.metrics import accuracy_score
print(f'Accuracy Score :{accuracy_score(y_test,lr.predict(X_test))}')
````

## Question 2: How many iterations did the algorithm take to converge?

````python
print(f'Number of iterations: {lr.n_iter_}')
````

Common instruction for Question 3 and 4


---

```
Perform Hyperparameter tuning on a decision tree classifier with random_state = 1 and cross validation = 4. The Hyperparameter tuning is to be done over the following parameters:

criterion can be 'entropy' or 'gini'

splitter can be 'random' or 'best'

Minimum number of samples per leaf as [2,4,6,8,10]

Maximum depth as [3,4,5,6]
```

##Question 3: What is the accuracy of the model on the test data? [Enter 3 decimal places]

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
dt = DecisionTreeClassifier(random_state=1)
param = {
    'criterion' : ['entropy', 'gini'],
    'splitter' : ['random', 'best'],
    'min_samples_leaf' : [2,4,6,8,10],
    'max_depth': [3,4,5,6]
}
cv = GridSearchCV(dt,param,cv=4,scoring='accuracy')
cv.fit(X_train,y_train)
````

````python
print(f'Accuracy Score: {accuracy_score(y_test,cv.best_estimator_.predict(X_test))}')
````

## Question 4: What is the value of best max_depth after training with GridSearchCV?

````python
cv.best_params_
````

````python
print(f'Value of best max_depth is: {cv.best_params_["max_depth"]}')
````

Common instruction for Question 3 and 4


---

```
Perform Hyperparameter tuning on an adaboost classifier with random_state = 1 and cross validation = 4. The Hyperparameter tuning is to be done over the following parameters:

Number of estimators as [100, 500, 1000]

Learning rate as [0.5, 1, 2]
```

##Question 5: What is the accuracy of the model on the test data? [Enter 3 decimal places]

````python
from sklearn.ensemble import AdaBoostClassifier
param_grid = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.5, 1, 2]
}
ada = AdaBoostClassifier(random_state=1)
adacv = GridSearchCV(ada,param_grid,cv=4,scoring='accuracy')
adacv.fit(X_train,y_train)
````

````python
print(f'Accuracy Score: {accuracy_score(y_test,adacv.best_estimator_.predict(X_test))}')
````

##Question 6: What is the value of best n_estimators after training with GridSearchCV?

````python
adacv.best_params_
````

````python
print(f'The best value of n_estimators is: {adacv.best_params_["n_estimators"]}')
````

Train a `voting classifier` by making use of the following estimators: `Bagging Classifier`, `Random Forest Classifier`, `GradientBoosting Classifier` with `random_state = 1` for each of the estimators.


Train the model on the training data and compute the score on the test data.

##Question 7: What is the accuracy of the model on the test data? [Enter 3 decimal places]


````python
from sklearn.ensemble import VotingClassifier, BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier
model_list = [
    ('bagg', BaggingClassifier(random_state=1)),
    ('rf',RandomForestClassifier(random_state=1)),
    ('gb', GradientBoostingClassifier(random_state=1))
]
vc = VotingClassifier(model_list)
vc.fit(X_train, y_train)
````

````python
print(f'Accuracy Score: {accuracy_score(y_test, vc.predict(X_test))}')
````
